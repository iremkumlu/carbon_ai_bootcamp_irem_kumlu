```python
from transformers import PegasusTokenizer, PegasusForConditionalGeneration
```

```python
# Model ve tokenizer yükleyelim
def load_pegasus_tokenizer():
    return PegasusTokenizer.from_pretrained("google/pegasus-xsum")
```

```python
def load_pegasus_model():
    return PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")
```

```python
# Özetleme fonksiyonu : 
def pegasus_summarize(text, tokenizer, model):
    # Metni modele uygun hale getirelim
    inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)

    # Modeli çalıştıralım
    summary_ids = model.generate(
        inputs['input_ids'], max_length=150, min_length=40, do_sample=True,
        temperature=1.5, top_k=50, top_p=0.95, repetition_penalty=1.2,
        length_penalty=2.0, num_return_sequences=1
    )

    # Özet metnini çözümlenmesi burada yapılır
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
# Tokenizer ve Model'i yükleyelim
tokenizer = load_pegasus_tokenizer()
model = load_pegasus_model()
```

```python
# Örnek input dosyamızı okuyalım
file_name = 'sample_input.txt'  
try:
    with open(file_name, 'r', encoding='utf-8') as file:  
        sample_text = file.read()
        print("Dosya başarıyla okundu.")
except FileNotFoundError:
    print(f"{file_name} adlı dosya bulunamadı. Lütfen dosya adını ve yolunu kontrol edin.")
    sample_text = None
```

```python
# Dosya başarıyla okunduysa Özetleme işlemi gerçekleşir
if sample_text:
    summary = pegasus_summarize(sample_text, tokenizer, model)
    print("Özetleme işlemi tamamlandı.")
    print("Özet:", summary)
```

```python
# Son olarak özetlenen metni bir dosyaya kaydedelim
output_file_name = "output_task1.txt"
try:
    with open(output_file_name, 'w', encoding='utf-8') as output_file:
        output_file.write(summary)
        print(f"Özetlenen metin '{output_file_name}' dosyasına kaydedildi.")
except Exception as e:
    print(f"Dosya kaydedilirken bir hata oluştu: {e}")
```
